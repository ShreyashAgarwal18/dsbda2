```python
# =========================================================
# DATA ANALYTICS I - LINEAR REGRESSION
# USING CALIFORNIA HOUSING DATASET (REPLACEMENT OF BOSTON)
# =========================================================

import pandas as pd

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# =========================================================
# 1. LOAD DATASET
# =========================================================

housing = fetch_california_housing()

df = pd.DataFrame(housing.data, columns=housing.feature_names)
df["PRICE"] = housing.target

print("Dataset Loaded Successfully")
print(df.head())

# =========================================================
# 2. SIMPLE LINEAR REGRESSION
# (Using one feature: MedInc)
# =========================================================

X_simple = df[["MedInc"]]
y = df["PRICE"]

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
    X_simple, y, test_size=0.3, random_state=42
)

model_simple = LinearRegression()
model_simple.fit(X_train_s, y_train_s)

y_pred_s = model_simple.predict(X_test_s)

mse_simple = mean_squared_error(y_test_s, y_pred_s)

print("\nSimple Linear Regression MSE:", mse_simple)

# =========================================================
# 3. MULTIPLE LINEAR REGRESSION
# =========================================================

X_multi = df.drop("PRICE", axis=1)

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(
    X_multi, y, test_size=0.3, random_state=42
)

model_multi = LinearRegression()
model_multi.fit(X_train_m, y_train_m)

y_pred_m = model_multi.predict(X_test_m)

mse_multi = mean_squared_error(y_test_m, y_pred_m)

print("\nMultiple Linear Regression MSE:", mse_multi)

# =========================================================
# 4. COMPARISON
# =========================================================

print("\n================ COMPARISON ================\n")

print("Simple Linear Regression MSE :", mse_simple)
print("Multiple Linear Regression MSE:", mse_multi)

if mse_multi < mse_simple:
    print("\nConclusion: Multiple Linear Regression performs better (lower error).")
else:
    print("\nConclusion: Simple Linear Regression performs better (lower error).")
```

```python
# =========================================================
# LINEAR REGRESSION - WITH MISSING VALUE HANDLING
# =========================================================

import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# =========================================================
# 1. LOAD DATASET
# =========================================================

path = r"C:\Users\Subhi Jain\Desktop\dsbda-prac_227\datasets\BostonHousing.csv"

df = pd.read_csv(path)

print("Dataset Loaded Successfully\n")
print(df.head())

# =========================================================
# 2. HANDLE MISSING VALUES (IMPORTANT FIX)
# =========================================================

print("\nChecking missing values:\n")
print(df.isnull().sum())

# Fill missing values with column mean
df = df.fillna(df.mean(numeric_only=True))

print("\nMissing values handled successfully.\n")

# =========================================================
# 3. TARGET COLUMN
# =========================================================

target = "medv"

# =========================================================
# 4. SIMPLE LINEAR REGRESSION
# =========================================================

X_simple = df[["lstat"]]
y = df[target]

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
    X_simple, y, test_size=0.3, random_state=42
)

model_simple = LinearRegression()
model_simple.fit(X_train_s, y_train_s)

y_pred_s = model_simple.predict(X_test_s)

mse_simple = mean_squared_error(y_test_s, y_pred_s)

print("Simple Linear Regression MSE:", mse_simple)

# =========================================================
# 5. MULTIPLE LINEAR REGRESSION
# =========================================================

X_multi = df.drop(target, axis=1)

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(
    X_multi, y, test_size=0.3, random_state=42
)

model_multi = LinearRegression()
model_multi.fit(X_train_m, y_train_m)

y_pred_m = model_multi.predict(X_test_m)

mse_multi = mean_squared_error(y_test_m, y_pred_m)

print("Multiple Linear Regression MSE:", mse_multi)

# =========================================================
# 6. COMPARISON
# =========================================================

print("\n================ COMPARISON ================\n")

print("Simple Linear Regression MSE :", mse_simple)
print("Multiple Linear Regression MSE:", mse_multi)

if mse_multi < mse_simple:
    print("\nConclusion: Multiple Linear Regression performs better.")
else:
    print("\nConclusion: Simple Linear Regression performs better.")
```
