```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# =========================================================
# LOAD DATASET
# =========================================================

path = r"C:\Users\Subhi Jain\Desktop\dsbda-prac_227\datasets\gdp.csv"
df = pd.read_csv(path)

print("\n================ DATA PREVIEW ================\n")
print(df.head())

# =========================================================
# COLUMN NAMES
# =========================================================

country_col = "country"
year_col = "year"
gdp_col = "gdpPercap"
```

```python
#a 
belgium = df[(df[country_col] == "Belgium") &
             (df[year_col] >= 2006) &
             (df[year_col] <= 2026)]

plt.figure(figsize=(10,5))
sns.barplot(data=belgium, x=year_col, y=gdp_col)
plt.title("GDP per Capita of Belgium (2006–2026)")
plt.xticks(rotation=45)
plt.show()

#b
countries = ["India", "Nepal", "Romania", "Singapore"]

gdp_2010 = df[(df[year_col] == 2010) &
              (df[country_col].isin(countries))]

plt.figure()
plt.pie(gdp_2010[gdp_col],
        labels=gdp_2010[country_col],
        autopct="%1.1f%%")

plt.title("GDP Distribution (2010)")
plt.show()

# =========================================================
# (c) BHUTAN VS COSTA RICA
# =========================================================

compare = df[df[country_col].isin(["Bhutan", "Costa Rica"])]

plt.figure(figsize=(8,5))
sns.lineplot(data=compare,
             x=year_col,
             y=gdp_col,
             hue=country_col,
             marker="o")

plt.title("GDP Comparison: Bhutan vs Costa Rica")
plt.show()

# =========================================================
#d
countries_scatter = ["Mexico", "Algeria", "Fiji", "Estonia"]

scatter_df = df[(df[country_col].isin(countries_scatter)) &
                (df[year_col] >= 2004) &
                (df[year_col] <= 2006)]

plt.figure(figsize=(10,6))
sns.scatterplot(data=scatter_df,
                x=year_col,
                y=gdp_col,
                hue=country_col,
                s=100)

plt.title("GDP Scatter Plot (2004–2006)")
plt.show()
```

```python
# =========================================================
# PRACTICAL 27 - DATA VISUALIZATION (GDP DATASET)
# FINAL ERROR-FREE VERSION
# =========================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# =========================================================
# LOAD DATASET
# =========================================================

path = r"C:\Users\Subhi Jain\Desktop\dsbda-prac_227\datasets\gdp.csv"
df = pd.read_csv(path)

print("\n================ DATA PREVIEW ================\n")
print(df.head())

print("\n================ COLUMNS ================\n")
print(df.columns)

# =========================================================
# COLUMN DEFINITIONS
# =========================================================

country_col = "country"
year_col = "year"
gdp_col = "gdpPercap"

# =========================================================
# CLEAN DATA (IMPORTANT)
# =========================================================

df = df.dropna(subset=[gdp_col])

# =========================================================
# (a) BAR GRAPH - BELGIUM (2006–2026)
# =========================================================

belgium = df[(df[country_col] == "Belgium") &
             (df[year_col] >= 2006) &
             (df[year_col] <= 2026)]

plt.figure(figsize=(10,5))
sns.barplot(data=belgium, x=year_col, y=gdp_col)
plt.title("GDP per Capita of Belgium (2006–2026)")
plt.xticks(rotation=45)
plt.show()

# =========================================================
# (b) PIE CHART - FIXED SAFE VERSION
# =========================================================

countries_pie = ["India", "Nepal", "Romania", "Singapore"]

gdp_2010 = df[(df[year_col] == 2010) &
              (df[country_col].isin(countries_pie))]

gdp_2010 = gdp_2010.dropna(subset=[gdp_col])

print("\nPIE DATA:\n", gdp_2010[[country_col, gdp_col]])

if not gdp_2010.empty:
    plt.figure(figsize=(6,6))
    plt.pie(gdp_2010[gdp_col],
            labels=gdp_2010[country_col],
            autopct="%1.1f%%")
    plt.title("GDP Distribution (2010)")
    plt.show()
else:
    print("No data available for Pie Chart (2010)")

# =========================================================
# (c) BHUTAN VS COSTA RICA
# =========================================================

compare = df[df[country_col].isin(["Bhutan", "Costa Rica"])]

plt.figure(figsize=(8,5))
sns.lineplot(data=compare,
             x=year_col,
             y=gdp_col,
             hue=country_col,
             marker="o")

plt.title("GDP Comparison: Bhutan vs Costa Rica")
plt.show()

# =========================================================
# (d) SCATTER PLOT (2004–2006)
# =========================================================

countries_scatter = ["Mexico", "Algeria", "Fiji", "Estonia"]

scatter_df = df[(df[country_col].isin(countries_scatter)) &
                (df[year_col] >= 2004) &
                (df[year_col] <= 2006)]

plt.figure(figsize=(10,6))
sns.scatterplot(data=scatter_df,
                x=year_col,
                y=gdp_col,
                hue=country_col,
                s=100)

plt.title("GDP Scatter Plot (2004–2006)")
plt.show()
```
