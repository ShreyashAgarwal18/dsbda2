```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# =========================================================
# LOAD DATASET
# =========================================================

path = r"C:\Users\Subhi Jain\Desktop\dsbda-prac_227\datasets\death_rate.csv"
df = pd.read_csv(path)

print("\n================ DATA PREVIEW ================\n")
print(df.head())

# =========================================================
# DEFINE COLUMNS
# =========================================================

year_col = "Year"
death_col = "Number_of_Deaths"

# =========================================================
# (a) VISUALIZATION
# =========================================================

# 1. Line Plot - Year wise deaths
plt.figure()
sns.lineplot(data=df, x=year_col, y=death_col)
plt.title("Year-wise Death Trend")
plt.show()

# 2. Bar Plot - Year wise comparison
plt.figure(figsize=(10,5))
sns.barplot(data=df, x=year_col, y=death_col)
plt.xticks(rotation=45)
plt.title("Year-wise Death Comparison")
plt.show()

# 3. Histogram - Distribution of deaths
plt.figure()
sns.histplot(df[death_col], kde=True)
plt.title("Distribution of Deaths")
plt.show()

# =========================================================
# (b) YEAR WITH MAXIMUM DEATHS
# =========================================================

max_row = df.loc[df[death_col].idxmax()]

print("\n================ MAXIMUM DEATH YEAR ================\n")
print("Year:", max_row[year_col])
print("Maximum Deaths:", max_row[death_col])

# =========================================================
# (c) COUNTRY-WISE DEATH COUNT
# =========================================================

country_data = df.groupby("Country")[death_col].sum().reset_index()

print("\n================ COUNTRY-WISE DEATH COUNT ================\n")
print(country_data)

plt.figure(figsize=(10,5))
sns.barplot(data=country_data, x="Country", y=death_col)
plt.xticks(rotation=45)
plt.title("Country-wise Death Count")
plt.show()
```
