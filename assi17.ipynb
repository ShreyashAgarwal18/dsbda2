```python
# =========================================================
# DATA VISUALIZATION II - TITANIC DATASET
# BOX PLOT: AGE vs GENDER + SURVIVAL STATUS
# =========================================================

import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('titanic')

print("Dataset Loaded")
print(df.head())

# =========================================================
# 1. BOXPLOT: AGE vs SEX with SURVIVAL INFO
# =========================================================

plt.figure(figsize=(8,5))

sns.boxplot(
    x='sex',
    y='age',
    hue='survived',
    data=df
)

plt.title("Age Distribution by Gender and Survival Status")
plt.xlabel("Gender (Sex)")
plt.ylabel("Age")
plt.legend(title="Survived (0=No, 1=Yes)")
plt.show()

# =========================================================
# 2. OBSERVATIONS (PRINTED FOR LAB RECORD)
# =========================================================

print("\n================ OBSERVATIONS ================\n")

print("1. Female passengers generally have higher survival rate than males.")
print("2. Male passengers show a wider age spread among non-survivors.")
print("3. Younger passengers (children) have higher survival tendency.")
print("4. Median age of survivors is slightly lower than non-survivors.")
print("5. Outliers exist in both male and female age distributions, especially in older age group.")

print("\n==============================================")
```
