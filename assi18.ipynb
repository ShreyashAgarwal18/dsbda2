```python
# =========================================================
# DATA VISUALIZATION III - IRIS DATASET (FIXED VERSION)
# =========================================================

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math

# Load dataset (update path if needed)
path = r"C:\Users\Subhi Jain\Desktop\dsbda-prac_227\datasets\Iris.csv"
df = pd.read_csv(path)

print("Dataset Loaded\n")
print(df.head())

# =========================================================
# 1. FEATURES AND TYPES
# =========================================================

print("\n================ 1. FEATURES & TYPES ================\n")

for col in df.columns:
    if df[col].dtype == "object":
        print(f"{col} : Nominal (Categorical)")
    else:
        print(f"{col} : Numeric (Continuous)")

# =========================================================
# 2. DISTRIBUTION (HIST + KDE) - FIXED
# =========================================================

print("\n================ 2. DISTRIBUTION (HIST + KDE) ================\n")

features = df.select_dtypes(include="number").columns
n = len(features)

cols = 2
rows = math.ceil(n / cols)

plt.figure(figsize=(12, 8))

for i, col in enumerate(features):
    plt.subplot(rows, cols, i + 1)
    sns.histplot(df[col], kde=True, bins=15)
    plt.title(f"Distribution of {col}")

plt.tight_layout()
plt.show()

# =========================================================
# 3. BOXPLOTS
# =========================================================

print("\n================ 3. BOXPLOTS ================\n")

plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title("Boxplot of Iris Features")
plt.show()

# =========================================================
# 4. COMPARISON USING STATISTICS
# =========================================================

print("\n================ 4. DISTRIBUTION COMPARISON ================\n")

print(df.describe().T[["mean", "std", "min", "max"]])

print("\nINFERENCES:")
print("- Petal length and width show clear clustering.")
print("- Sepal width shows highest variability and some outliers.")
print("- Sepal length is moderately distributed.")
print("- Petal features are most useful for classification.")
```
