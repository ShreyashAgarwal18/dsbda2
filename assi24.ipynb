```python
import pandas as pd

# =========================================================
# 1. LOAD EXISTING DATASET
# =========================================================

path = r"C:\Users\Subhi Jain\Desktop\dsbda-prac_227\datasets\Food_Delivery_Times.csv"

df = pd.read_csv(path)

print("\n================ DATASET PREVIEW ================\n")
print(df.head())

# =========================================================
# 2. CHECK COLUMNS
# =========================================================

print("\nColumns in dataset:\n", df.columns)

# Assume dataset has:
# City (categorical)
# Delivery_Time (numeric)

# If column names differ, adjust here:
category_col = df.columns[0]
numeric_col = df.columns[1]

# =========================================================
# 3. GROUPED STATISTICS
# =========================================================

print("\n================ GROUPED STATISTICS ================\n")

stats = df.groupby(category_col)[numeric_col].agg([
    "mean",
    "median",
    "min",
    "max",
    "std"
])

print(stats)

# =========================================================
# 4. LIST OF VALUES PER CATEGORY
# =========================================================

print("\n================ LIST OF VALUES PER CATEGORY ================\n")

group_list = df.groupby(category_col)[numeric_col].apply(list)

print(group_list)
```

```python
# =========================================================
# IRIS DATASET - BASIC STATISTICAL DETAILS BY SPECIES
# =========================================================

import pandas as pd
import numpy as np

# =========================================================
# 1. LOAD EXISTING IRIS DATASET
# =========================================================

path = r"C:\Users\Subhi Jain\Desktop\dsbda-prac_227\datasets\Iris.csv"
df = pd.read_csv(path)

print("\n================ IRIS DATASET ================\n")
print(df.head())

# detect species column
species_col = df.columns[-1]

# numeric columns only
num_cols = df.select_dtypes(include=np.number).columns

# =========================================================
# 2. FUNCTION TO COMPUTE STATISTICS
# =========================================================

def species_stats(species_name):
    data = df[df[species_col] == species_name][num_cols]
    
    print(f"\n================ {species_name} =================\n")
    
    for col in num_cols:
        values = data[col]
        
        print(f"\nFeature: {col}")
        print("Mean:", values.mean())
        print("Standard Deviation:", values.std())
        print("25th Percentile:", np.percentile(values, 25))
        print("50th Percentile (Median):", np.percentile(values, 50))
        print("75th Percentile:", np.percentile(values, 75))

# =========================================================
# 3. APPLY FOR SPECIES
# =========================================================

species_stats("Iris-setosa")
species_stats("Iris-versicolor")
species_stats("Iris-virginica")
```
