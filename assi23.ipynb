```python
import pandas as pd
import numpy as np

# =========================================================
# 1. CREATE DATAFRAME
# =========================================================

data = [
    [1, "Asha", "F", 85, 78, 88, 92],
    [2, "Rahul", "M", 90, 85, None, 85],
    [3, "Priya", "F", 78, 72, 70, 88],
    [4, "Arjun", "M", 120, 88, 92, 95],   # outlier
    [5, "Kavya", "F", 65, 60, 58, None],
    [6, "Rohan", "M", 55, 48, 52, 80],
    [7, "Neha", "F", 95, 98, 100, 97],
    [8, "Amit", "M", 40, 42, 38, 70],
    [9, "Sneha", "F", 88, 85, 90, "abc"],  # invalid value
    [10, "Varun", "M", 72, 75, 70, 85]
]

columns = ["RollNo", "Name", "Gender", "Maths", "Science", "English", "Attendance"]

df = pd.DataFrame(data, columns=columns)

print("\nORIGINAL DATASET:\n")
print(df)

# =========================================================
# 2. CLEAN DATA (HANDLE MISSING + INVALID VALUES)
# =========================================================

# Convert invalid values to NaN
df["Attendance"] = pd.to_numeric(df["Attendance"], errors="coerce")

# Convert other numeric columns safely
for col in ["Maths", "Science", "English", "Attendance"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

print("\nMISSING VALUES BEFORE:\n")
print(df.isnull().sum())

# Fill missing values with column MEAN
df["English"].fillna(df["English"].mean(), inplace=True)
df["Attendance"].fillna(df["Attendance"].mean(), inplace=True)

print("\nAFTER MISSING VALUE HANDLING:\n")
print(df)

# =========================================================
# 3. OUTLIER HANDLING (IQR METHOD)
# =========================================================

def remove_outliers(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    median = df[col].median()

    df[col] = np.where((df[col] < lower) | (df[col] > upper), median, df[col])

for col in ["Maths", "Science", "English", "Attendance"]:
    remove_outliers(col)

print("\nAFTER OUTLIER HANDLING:\n")
print(df)

# =========================================================
# 4. NORMALIZATION (0 TO 1 SCALING)
# =========================================================

df["Maths_Normalized"] = (df["Maths"] - df["Maths"].min()) / (df["Maths"].max() - df["Maths"].min())

print("\nFINAL DATASET WITH NORMALIZATION:\n")
print(df)
```
