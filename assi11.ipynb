```python
# ==============================
# TEXT ANALYTICS (SINGLE DOCUMENT)
# ==============================

import nltk
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag

# ==============================
# DOWNLOAD REQUIRED PACKAGES
# ==============================
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')

# ==============================
# SINGLE DOCUMENT
# ==============================
doc = """
Python is widely used in data science and machine learning.
It helps in building intelligent systems and real-world applications.
"""

print("Original Document:\n", doc)

# ==============================
# TOKENIZATION
# ==============================
tokens = word_tokenize(doc.lower())
print("\nTokenization:\n", tokens)

# ==============================
# STOPWORD REMOVAL
# ==============================
stop_words = set(stopwords.words('english'))

filtered_tokens = [
    word for word in tokens
    if word.isalnum() and word not in stop_words
]

print("\nAfter Stopword Removal:\n", filtered_tokens)

# ==============================
# POS TAGGING
# ==============================
pos_tags = pos_tag(filtered_tokens)
print("\nPOS Tagging:\n", pos_tags)

# ==============================
# STEMMING
# ==============================
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]

print("\nStemming:\n", stemmed_words)

# ==============================
# LEMMATIZATION
# ==============================
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("\nLemmatization:\n", lemmatized_words)

# ==============================
# TERM FREQUENCY (TF)
# ==============================
cv = CountVectorizer()
tf_matrix = cv.fit_transform([doc])

tf_df = pd.DataFrame(tf_matrix.toarray(), columns=cv.get_feature_names_out())

print("\nTerm Frequency (TF):\n", tf_df)

# ==============================
# INVERSE DOCUMENT FREQUENCY (IDF)
# ==============================
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform([doc])

idf_df = pd.DataFrame(
    tfidf.idf_,
    index=tfidf.get_feature_names_out(),
    columns=["IDF"]
)

print("\nInverse Document Frequency (IDF):\n", idf_df)
```
