```python
# =========================================================
# DATA ANALYTICS III - NAIVE BAYES CLASSIFICATION
# IRIS DATASET
# =========================================================

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

# =========================================================
# 1. LOAD DATASET (DOWNLOADED CSV)
# =========================================================

path = r"C:\Users\Subhi Jain\Desktop\dsbda-prac_227\datasets\Iris.csv"

df = pd.read_csv(path)

print("Dataset Loaded Successfully\n")
print(df.head())

# =========================================================
# 2. SPLIT FEATURES AND TARGET
# =========================================================

X = df.drop(df.columns[-1], axis=1)   # all columns except last
y = df[df.columns[-1]]                # last column = species

# =========================================================
# 3. TRAIN-TEST SPLIT
# =========================================================

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# =========================================================
# 4. NAIVE BAYES MODEL
# =========================================================

model = GaussianNB()
model.fit(X_train, y_train)

# =========================================================
# 5. PREDICTION
# =========================================================

y_pred = model.predict(X_test)

# =========================================================
# 6. CONFUSION MATRIX
# =========================================================

cm = confusion_matrix(y_test, y_pred)

print("\nConfusion Matrix:")
print(cm)

# =========================================================
# 7. TP, FP, TN, FN (MULTICLASS HANDLING)
# =========================================================

FP = cm.sum(axis=0) - np.diag(cm)
FN = cm.sum(axis=1) - np.diag(cm)
TP = np.diag(cm)
TN = cm.sum() - (FP + FN + TP)

print("\nTP:", TP)
print("FP:", FP)
print("FN:", FN)
print("TN:", TN)

# =========================================================
# 8. METRICS
# =========================================================

accuracy = accuracy_score(y_test, y_pred)
error_rate = 1 - accuracy
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')

print("\nAccuracy:", accuracy)
print("Error Rate:", error_rate)
print("Precision:", precision)
print("Recall:", recall)
```
