```python
# =========================================================
# DESCRIPTIVE STATISTICS - MEASURES OF CENTRAL TENDENCY
# AND VARIABILITY (WINE + IRIS DATASET)
# =========================================================

import pandas as pd
import numpy as np

# =========================================================
# 1. WINE DATASET - SUMMARY STATISTICS GROUPED BY CATEGORY
# =========================================================

# Full path (update if needed)
path_wine = r"C:\Users\Subhi Jain\Desktop\dsbda-prac_227\datasets\winequality-red.csv"

df_wine = pd.read_csv(path_wine)

print("\nWINE DATASET HEAD:")
print(df_wine.head())

# Assume dataset has a categorical column (e.g., "quality")
# and numeric columns like alcohol, acidity, etc.

# If your dataset uses different column name, change here:
categorical_col = "quality"

numeric_cols = df_wine.select_dtypes(include=np.number).columns

# Group by categorical column and calculate statistics
grouped_stats = df_wine.groupby(categorical_col)[numeric_cols].agg(
    ['mean', 'median', 'min', 'max', 'std']
)

print("\nSUMMARY STATISTICS (GROUPED BY QUALITY):")
print(grouped_stats)
```

```python
# =========================================================
# 2. IRIS DATASET - BASIC STATISTICAL DETAILS
# =========================================================

path_iris = r"C:\Users\Subhi Jain\Desktop\dsbda-prac_227\datasets\Iris.csv"

df_iris = pd.read_csv(path_iris)

print("\nIRIS DATASET HEAD:")
print(df_iris.head())

# Assuming last column is species
species_col = df_iris.columns[-1]

# Numeric columns only
iris_numeric = df_iris.select_dtypes(include=np.number)

# Function to calculate stats for each species
def species_stats(species_name):
    subset = df_iris[df_iris[species_col] == species_name]
    
    print("\n===================================")
    print("SPECIES:", species_name)
    print("===================================")
    
    print("\nMean:\n", subset[iris_numeric.columns].mean())
    print("\nMedian:\n", subset[iris_numeric.columns].median())
    print("\nPercentiles (25%, 50%, 75%):\n", 
          subset[iris_numeric.columns].quantile([0.25, 0.5, 0.75]))
    print("\nStandard Deviation:\n", subset[iris_numeric.columns].std())

# Apply for all 3 species
species_stats("Iris-setosa")
species_stats("Iris-versicolor")
species_stats("Iris-virginica")
```
